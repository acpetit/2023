```python
def getgalaxies(observation,redshift=2):
    expandedcoordinatesx = np.cumsum(1+(redshift-1)*(observation.sum(axis=0)==0))
    expandedcoordinatesy = np.cumsum(1+(redshift-1)*(observation.sum(axis=1)==0))
    galaxies = []
    for y,line in enumerate(observation):
        for x,pixel in enumerate(line):
            if pixel:
                galaxies.append((expandedcoordinatesx[x],expandedcoordinatesy[y]))
    return galaxies

def galaxydist(galaxies):

    sumofpaths = 0

    for (g1,g2) in itertools.combinations(galaxies,2):
        sumofpaths += abs(g1[0]-g2[0])+abs(g1[1]-g2[1])
     

    return sumofpaths
```

# Part 1

```python
galaxiesz2= getgalaxies(cosmodata,redshift=2)
print(f"Sum of shortest paths: {galaxydist(galaxiesz2)}")

```
