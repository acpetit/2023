# Part 1

```python
with open("input25.txt") as f:
    wiremap ={}
    for row in f:
        k,v = row.rstrip().split(": ")
        wiremap[k] = [k for k in v.split()]
wgraph= nx.Graph(wiremap)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
nx.draw(wgraph, with_labels=True,node_size=20,ax=ax)
```
